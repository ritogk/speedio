```python
import googlemaps
import os
from dotenv import load_dotenv
load_dotenv()

# クライアントの作成
gmaps = googlemaps.Client(key=os.environ['GOOGLE_API_KEY'])
# 元の緯度経度
original_location = (35.466453, 136.938538)
# 512件の同じ位置情報を持つリストを作成
locations = [original_location] * 712

elevations = []
# 1度のリクエストで512箇所までしか指定できないので分割してリクエストを送る
for i in range(0, len(locations), 512):
    print(1)
    elevations.extend(gmaps.elevation(locations[i:i+512]))

print(len(elevations))

# 結果の表示
for elevation_data in elevations:
    location = elevation_data['location']
    elevation = elevation_data['elevation']
    print(f"緯度: {location['lat']}, 経度: {location['lng']}, 標高: {elevation} メートル")
```
