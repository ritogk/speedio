```python
import osmnx as ox
import folium

# 地域の範囲を指定（例：東京都渋谷区）
place_name = "渋谷区, 東京都, Japan"

# 大通りの道を抽出
graph = ox.graph_from_point(center_point=(35.3540567, 136.9216636), network_type='drive', dist=500)

# 大通りの道のみを抽出
primary_streets = ox.utils_graph.get_largest_component(graph, strongly=True)

ox.save_graphml(graph,filepath='osaka_all.graphml')

# 大通りの道の地図を描画
map_osm = ox.plot_graph_folium(primary_streets, edge_width=2)

# 地図を表示
map_osm
```
